```python
import pandas as pd
# from sklearn.model_selection import LogisticRegression 
```

```python
df=pd.read_csv(r"C:\python code\machine learning\17_l1&l2_regularization\Melbourne_housing_FULL.csv")
df.head()
```

```python
df.drop(['Date','CouncilArea','Lattitude','Longtitude','Regionname'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
clos_to_fill_zero=['Propertycount','Distance','Bathroom','Bedroom2', 'Car']
df[clos_to_fill_zero]=df[clos_to_fill_zero].fillna(0)
df.head()
```

```python
df.isna().sum()
```

```python
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
X=df.drop(['Price'],axis=1)
y=df.Price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn import linear_model
reg=linear_model()
```
