```python
le=LabelEncoder()
ledf=df
```

```python
ledf['Car Model']=le.fit_transform(ledf['Car Model'])
ledf
```

```python
X=ledf[['Car Model',	'Mileage',	'Age(yrs)']].values
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming you want to apply OneHotEncoder to the first column only
ohe = ColumnTransformer(
    transformers=[('Car Model', OneHotEncoder(sparse_output=False), [0])],
    remainder='passthrough'
)
```

```python
X=ohe.fit_transform(X)
X
```

```python
X=X[:,1:]
```

```python
X
```

```python
Y=ledf['Sell Price($)'].values
Y
```

```python
lin=LinearRegression()
```

```python
lin.fit(X,Y)
```

```python
lin.predict([[1.00e+00, 0.00e+00, 6.90e+04, 6.00e+00]])
```

```python
lin.score(X,Y)
```
