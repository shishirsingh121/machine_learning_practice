```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
df=pd.read_csv(r"C:\python code\machine learning\14_naive_bays_classifier\titanic.csv")
df.head()
```

```python
df.Sex=df.Sex.map({'male':0,'female':1})
df
```

```python
inputs=df[['Pclass','Sex','Age','Fare']]
inputs
```

```python
df.target=df.Survived
df.target
```

```python
inputs.info()
```

```python
inputs.isna().sum()
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
inputs.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,df.target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
