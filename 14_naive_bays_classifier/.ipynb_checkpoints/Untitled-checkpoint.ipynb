```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
df=pd.read_csv(r"C:\python code\machine learning\14_naive_bays_classifier\titanic.csv")
df.head()
```

```python
dummies=pd.get_dummies(df.Sex,df.Sex.dtype=='int64')
dummies.head()
```

```python
input=df[['Pclass','Sex',	'Age',			'Fare'		]]
input
```

```python
target=df.Survived
target
```

```python
input=pd.concat([input,dummies],axis=1)
input
```

```python
input.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(input,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_test)
```

```python
model.score(X_train,y_test)
```
