```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.DataFrame({
    "area":[2300,2400,2500,2600,2700],
    "price":[23400,23600,23900,24000,24200]
})
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color='red',marker="+")
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)

```

```python
reg.predict([[2400]])
```
