```python
import pandas as pd
from sklearn import svm
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df

```

```python
df['target']=iris.target
df
```

```python
df['target_flowers']=df['target'].apply(lambda x: iris.target_names[x])
df
```

```python
from sklearn.model_selection import train_test_split 
X_trian, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
model=svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_trian,y_train)
model.score(X_test,y_test)
#ots change the score according to given data so its approch is not that use ful
```

```python
from  sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=30,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=30,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=50,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
kernel=['linear','rbf']
C=[10,20,50]
avg_score={}
for kval in kernel:
    for cval in C:
        C_score=cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),iris.data,iris.target,cv=5)
        avg_score[kval+'-'+str(cval)]=np.average(C_score)
avg_score        
        
```

```python
from sklearn.model_selection import GridSearchCV
clf=GridSearchCV(svm.SVC(gamma='auto'),{
    'kernel':['linear','rbf'],
    'C':[1,20,50]
},cv=5,return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_

```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_kernel','param_C','mean_test_score']]
```
